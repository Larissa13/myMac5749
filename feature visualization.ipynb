```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from matplotlib.colors import ListedColormap
from mpl_toolkits.mplot3d import axes3d, Axes3D
%matplotlib inline
```

```python
def featureSpaceVisualize(featureMatrix, nFeatures, nClassification):
    twodim  = PCA(n_components=2)
    threedim = PCA(n_components=3)
    twodim_class = twodim.fit_transform(featureMatrix[:, :nFeatures])
    threedim_class = threedim.fit_transform(featureMatrix[:, :nFeatures])
    m = np.max(featureMatrix[:,nFeatures:])
    colors = plt.cm.rainbow(np.linspace(0,1, m))
    for i in range(nClassification):
        
        class_i = featureMatrix[:,nFeatures+i]
        histogram = np.histogram(class_i, bins= np.arange(np.max(class_i)))
        plt.hist(class_i)
        plt.show()
        ncolors = np.unique(class_i).shape[0]
        cmap = ListedColormap(colors[:ncolors])
        plt.scatter(twodim_class[:, 0],twodim_class[:, 1], c=class_i, cmap=cmap)
        plt.show()
        fig = plt.figure()
        ax = Axes3D(fig)
        ax.scatter(threedim_class[:, 0], threedim_class[:, 1], threedim_class[:, 2], c=class_i, cmap=cmap)
        plt.show()
```

```python
a = np.array([[1.0, 2.0, 0, 1], [0.3, 0.2, 0, 3], [1.8, 9.8, 4, 1]])
featureSpaceVisualize(a, 3, 1)
```
